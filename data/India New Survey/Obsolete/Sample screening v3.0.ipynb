## Duration based screening

```python
# Dropping outliers
sns.boxplot(data = df1["Duration (in seconds)"])
```

```python
df2 = df1.reset_index()
sns.displot( data = df1["Duration (in seconds)"].astype(np.int64), kind = "kde")
```

```python
sns.scatterplot(data=df1, y=df1.index, x="Duration (in seconds)")
```

```python
# removing that two extreme outliers

dfFilter1 = df1.loc[ df1["Duration (in seconds)"] < 10000]
sns.scatterplot(data=dfFilter1, y=dfFilter1.index, x="Duration (in seconds)")
```

```python
sns.displot( data = dfFilter1["Duration (in seconds)"], kind = "kde")
```

```python
sns.boxplot(data = dfFilter1["Duration (in seconds)"])
```

```python
filter1Mean = dfFilter1["Duration (in seconds)"].mean()
filter1Std = dfFilter1["Duration (in seconds)"].std()
filter1Median = dfFilter1["Duration (in seconds)"].median()

print("Mean =",filter1Mean/60,"; SD =", filter1Std/60, "; Median =", filter1Median/60)

# we find that mean = median
```

```python
# Mean +/- 1SD

lowerThres1 = filter1Mean - filter1Std
higherThres1 = filter1Mean + filter1Std

print("Higher thres =",higherThres1/60,"; Lower thres =", lowerThres1/60)
```

```python
# using this to filter: 51 data points are excluded

dfFilter2 = dfFilter1.loc[ (dfFilter1["Duration (in seconds)"] < higherThres1) & (dfFilter1["Duration (in seconds)"] > lowerThres1) ].copy()
dfFilter2["Duration (in seconds)"].size - dfFilter1["Duration (in seconds)"].size
```

```python
# Median +/- 1SD

lowerThres2 = filter1Median - filter1Std
higherThres2 = filter1Median + filter1Std

print("Higher thres =",higherThres2/60,"; Lower thres =", lowerThres2/60)
```

```python
# using this to filter: 39 data points are excluded

dfFilter3 = dfFilter1.loc[ (dfFilter1["Duration (in seconds)"] < higherThres2) & (dfFilter1["Duration (in seconds)"] > lowerThres2) ].copy()
dfFilter3["Duration (in seconds)"].size - dfFilter1["Duration (in seconds)"].size
```
