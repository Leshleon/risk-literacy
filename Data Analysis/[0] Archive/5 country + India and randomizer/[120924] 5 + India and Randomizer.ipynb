```python
import pandas as pd
pd.set_option("display.max_columns", None)

from tqdm import tqdm
tqdm.pandas()

import numpy as np
```

```python
india = pd.read_csv("India_riskliteracy_dataset_above median_2024.csv")
india.drop(["Unnamed: 1", "#.1"], axis = 1).rename({"#":"responseid"}, axis = 1)
```

```python
otherCountries = pd.read_csv("5country_riskliteracy_dataset_2024.csv")
otherCountries = otherCountries.drop(["Unnamed: 1", "responseid.1"], axis = 1)
otherCountries
```

```python
fiveCountries = pd.concat([otherCountries,india])
```

```python
fiveCountries = fiveCountries.reset_index().rename({"index":"randomizerid"}, axis = 1)
fiveCountries
```

```python
fiveCountriesSample50_1 = fiveCountries.sample(frac = 0.5, replace = False).copy()
fiveCountriesSample50_1.sort_values(by = ["randomizerid"])
```

```python
fiveCountriesSample50_2 = fiveCountries.loc[ ~fiveCountries.index.isin(fiveCountriesSample50_1.index) ]
```

```python
fiveCountriesSample50_2.sort_values(by = ["randomizerid"])
```

```python
fiveCountriesSample50_1.to_csv("5_india_riskliteracy_2024_sample1.csv")
```

```python
fiveCountriesSample50_2.to_csv("5_india_riskliteracy_2024_sample2.csv")
```
