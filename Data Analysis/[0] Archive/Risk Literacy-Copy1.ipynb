```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Data Frame 1 or df1 is an aggregate data on all questions and facet total scores and over all total score vs each response

df1 = pd.read_excel('Survey Data.xlsx', sheet_name = 'Risk Facet scores')
df1
```

```python
# Data Frame 2 or df2 is an aggregate data on facet total scores and over all total score vs each response

df2 = pd.read_excel('Survey Data.xlsx', sheet_name = 'Wealth and Income')
df2
```

```python
# The following graph represents frequency of each data point on "Total Score_19" or the total score out of 19 across the sample.

df1['Total Score_19'].plot(kind = 'hist', xticks = np.arange(0, 20, step=1), xlabel = 'Total Score_19', title = 'Frequency of Scores' )
```

```python
# Data Frame 3 or df3 is an aggregate data on facet total scores and over all total score, along with wealth and income data vs each response. This data set is also sorted by wealth with all responses '7' removed (7 refers to "prefer not say" option).

df3 = df2.sort_values(by = 'Q12. Wealth')
df3 = df3.reset_index(drop = True)
df3.drop(df3[df3['Q12. Wealth'] == 7].index, inplace = True)
df3
```
